```python
arr1=np.array([
    [10,20,30],
    [30,40,50]
])
arr2=np.array([
    [10,20,30],
    [30,42,51]
])
print(arr1*arr2)
print(arr1/arr2)
print(arr1%arr2)
```

```python
#actual matrix multiplication
arr1=np.array([
    [10,20,30],
    [30,40,50]
])
arr2=np.array([
    [10,20],
    [30,42],
    [51,60]
])
print(arr1 @ arr2)
```

```python
#broadcasting
arr1=np.array([
    [10,20,30],
    [30,40,50],
    [30,42,51]
])
arr2=np.array([
    [10,20,30]
])
print(arr1*arr2)
print(arr1+5)
```

```python
arr1=np.array([
    [10,20,30],
    [30,40,50],
    [30,42,51]
])
arr3=arr1.flatten() # creates a new array
print(arr3)
arr4=arr1.ravel() # doesn't create a new array.

print(arr4)
```
